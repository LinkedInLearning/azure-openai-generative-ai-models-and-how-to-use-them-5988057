# Converting Text to speech using OpenAI TTS models

```python
import os
from openai import AzureOpenAI

# Setting up the client
client = AzureOpenAI(
    api_key=os.getenv("AZURE_OPENAI_API_KEY"),  
    api_version="2024-05-01-preview",
    azure_endpoint = os.getenv("AZURE_OPENAI_ENDPOINT")
)

# Define the Model
deployment_id = "tts"

response = client.audio.speech.create(
    model=deployment_id,
    voice="nova", # nova, shimmer, echo, onyx, fable, alloy
    input="Today is a wonderful day to build something amazing.",
    response_format="mp3"
)

# Play the audio
import IPython.display as ipd
ipd.Audio(data = response.read())
```
